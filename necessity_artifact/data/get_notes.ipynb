```python
import openpyxl
import pandas as pd

# Abre el archivo de Excel
file_path = 'lista_innovacion.xlsx'
workbook = openpyxl.load_workbook(file_path)

# Selecciona la hoja de trabajo
sheet = workbook.active

# Crear un DataFrame vacío con las mismas dimensiones que la hoja de Excel
data = pd.DataFrame(index=range(sheet.max_row), columns=range(sheet.max_column))

# Recorre todas las celdas de la hoja
for row in sheet.iter_rows():
    for cell in row:
        if cell.comment:
            # Guarda el comentario en la celda correspondiente del DataFrame
            data.iloc[cell.row - 1, cell.column - 1] = cell.comment.text

# Cierra el archivo
workbook.close()

# Muestra el DataFrame
print(data)
```

```python
data.to_excel('comentarios.xlsx', index=False)
```
